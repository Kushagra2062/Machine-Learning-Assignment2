```python
#For convenience we are defining a function called "compute_impurity()" that will calculate impurity of a feature using gini index
def compute_impurity(feature):
    probs = feature.value_counts(normalize=True)
    impurity = 1 - np.sum(np.square(probs))
    
    return round(impurity,3)
```

```python
def comp_feature_information_gain(df,target,descriptive_feature):
    print("Target feature:",target)
    print("Descriptive feature:",descriptive_feature)
    
    target_impurity=compute_impurity(df[target])
    #print(target_impurity)
    entropy_list=list()
    weight_list=list()
    
    for level in df[descriptive_feature].unique():
        df_feature_level = df[df[descriptive_feature] == level]
        entropy_level = compute_impurity(df_feature_level[target])
        entropy_list.append(round(entropy_level, 3))
        weight_level = len(df_feature_level) / len(df)
        weight_list.append(round(weight_level, 3))
        
    print("Impurity of partitions:",entropy_list)
    print("Weighs of partitions:",weight_list)
    
    feature_remaining_impurity = np.sum(np.array(entropy_list) * np.array(weight_list))
    print('Information impurity:', feature_remaining_impurity)
    
    information_gain = target_impurity - feature_remaining_impurity
    print("Information Gain:",information_gain)
    
    print("==========================")
    return information_gain
```

```python
def calc_info_gain(df, left, right, entropy_before_split, split_values):
    print("Split Value = {} \n".format(split_values))    
    total_elements = df.shape[0]
    print("Left Node")
    ent_left = compute_impurity(left)
    print("Entropy of Left Node {} \n".format(ent_left))
    print("Right Node")
    ent_right = compute_impurity(right)
    print("Entropy of Right Node {} \n".format(ent_right))
    weighted_entropy = ((left.shape[0] / total_elements) * ent_left) + (
        (right.shape[0] / total_elements) * ent_right
    )
    info_gain = entropy_before_split - weighted_entropy
    print("Info gain at split {} is {} \n".format(split_values, info_gain))
    return info_gain
```

```python
info_gains = []
for sp_value in split_values:
    left_data = a21.loc[a21.Age < sp_value]
    right_data = a21.loc[a21.Age > sp_value]
    info_gains.append(
        calc_info_gain(
            a21, left_data.Annual_Income, right_data.Annual_Income, impurity, sp_value
        )
    )
```
